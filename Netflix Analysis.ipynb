Changing date_added format

```python
nf['date_added'] = pd.to_datetime(nf['date_added'], format='%B %d, %Y', errors='coerce')
nf
```

To divide Duration in 2 parts  1)Duration in numbers (duration_num)  2)Duration in units(duration_unit)


```python
nf['duration_num']=nf['duration'].str.extract('(\d+)').astype(float)
nf['duration_unit']=nf['duration'].str.extract('([a-zA-Z]+)')
nf
```

To Filter By help of unit(minute) which are movies

```python
movies=nf[nf['duration_unit']=='min']
movies
```

To Filter By help of unit(Seasons) which are tv_show

```python
tv_show= nf[nf['duration_unit'].str.contains('Season',na=False)]
tv_show
```

Average Runtime of movies

```python
movies['duration_num'].mean()
```

Average Runtime of tv_show

```python
tv_show['duration_num'].mean()
```

Droping Missing Data From Dataset

```python
nf = nf.dropna(subset=['date_added', 'duration'])
nf
```

By filling Unkown Data in the Dataset

```python
nf=nf.copy()
nf.loc[:, 'director'] = nf['director'].fillna('Unknown')
nf.loc[:, 'cast'] = nf['cast'].fillna('Unknown')
nf.loc[:, 'country'] = nf['country'].fillna('Unknown')
nf.loc[:, 'rating'] = nf['rating'].fillna('Unknown')
nf
```

Comparison of Movies vs Tv Shows

```python
#Counts 
type_counts = nf['type'].value_counts()
type_counts
```

```python
#Graphical Representation
plt.pie(type_counts.values,labels=type_counts.index,autopct='%1.1f%%',startangle=140)
plt.title('Movies vs TV Shows on Netflix')
plt.show()
```

Content added per year

```python
#Adding year column through help of date_added
nf['year']=nf['date_added'].dt.year
nf
```

```python
#Counting of Adding movies and tv_shows each year
count_year=nf['year'].value_counts().sort_index()
count_year
```

```python
#Graphical Representation
plt.bar(count_year.index,count_year.values)
plt.xlabel('year Added')
plt.ylabel('Number of titles')
plt.title('Content added per year')
plt.grid(axis='y')
plt.show()
```

Top producing countries

```python
#remove unkown countries
allc = nf[nf['country']!='Unknown']['country'].str.split(', ').sum()
allc
```

```python
#count of each known country
couc=Counter(allc)
couc
```

```python
#listing top 10 common countries
comc=couc.most_common(10)
comc
```

```python
#Graphical Representation
countries, counts = zip(*comc)
plt.figure(figsize=(12, 6))
plt.bar(countries, counts)
plt.xlabel('Countries')
plt.ylabel('Number of Movies produced')
plt.title('Top 10 Countries')
plt.grid(axis='y')
plt.show()
```

Top 10 Genres

```python
#List of all Genres
allg=nf[nf['listed_in']!='Unknown']['listed_in'].str.split(', ').sum()
allg
```

```python
#Counting of Geners
countg=Counter(allg)
countg
```

```python
#Finding Top 10 Geners
topg=countg.most_common(10)
topg
```

```python
#Graphical Representation
gener, count = zip(*topg)
plt.figure(figsize=(15,8))
plt.bar(gener, count)
plt.xlabel('gener')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Counts')
plt.title('Top 10 Geners')
plt.grid(axis='y')
plt.show()
```

Heatmap â€” Titles by Country vs Year

```python
print("Top 10 countries:", countries)
df_top = nf[nf['country'].isin(countries)].copy()
heatmap_data = df_top.groupby(['country', 'year']).size().reset_index(name='count')
heatmap_pivot = heatmap_data.pivot(index='country', columns='year', values='count').fillna(0)
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_pivot, cmap='YlGnBu', annot=True, fmt='.0f')
plt.title('Titles by Country vs Year (Top 10 Countries)')
plt.xlabel('Year')
plt.ylabel('Country')
plt.show()
```

Movie Duration by Year

```python
#dropping null values for better understandings 
movies = nf[nf['type'] == 'Movie'].dropna(subset=['duration_num', 'release_year'])
movies
```

```python
#Graphical Representation
plt.figure(figsize=(16,6))
sns.boxplot(x='release_year', y='duration_num', data=movies)
plt.title('Movie Duration by Year')
plt.suptitle('')
plt.xlabel('Release Year')
plt.ylabel('Duration (min)')
plt.xticks(rotation=90)
plt.show()
```

Tv show Duration by Year

```python
#dropping null values for better understandings 
tv_show = nf[nf['type'] == 'TV Show'].dropna(subset=['duration_num', 'release_year'])
tv_show
```

```python
#Graphical Representation
plt.figure(figsize=(16,6))
sns.boxplot(x='release_year', y='duration_num', data=tv_show)
plt.title('Movie Duration by Year')
plt.suptitle('')
plt.xlabel('Release Year')
plt.ylabel('Duration (min)')
plt.xticks(rotation=90)
plt.show()
```

Genre Popularity Over Years

```python
#sorting of popularity
nf = nf.copy()
nf['main_genre'] = nf['listed_in'].str.split(',').str[0]
countsg = nf.groupby(['release_year', 'main_genre']).size().unstack(fill_value=0)
countsg = countsg.sort_index()
countsg
```

```python
#Graphical Reprsentation 
plt.figure(figsize=(16,8))
plt.stackplot(countsg.index,countsg.T.values, labels=countsg.columns)
plt.title('Genre Popularity Over Years (First Genre Only)')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.legend(title='Genre', bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
